```python
import pandas as pd
import numpy as np
```

```python
game=pd.read_csv('game.csv')
```

```python
# Convertir la columna 'fecha' a formato datetime
game['game_date'] = pd.to_datetime(game['game_date'])

#2018 en adelante
game_filtrada= game[game['game_date'] >  '2018-10-01']

# Verificar el resultado
game_filtrada
```

```python
print(game_filtrada.isnull().sum())
```

```python
# Crear una copia del DataFrame para evitar la advertencia
game_filtrada = game_filtrada.copy()
# Reemplazar los valores nulos en la columna 'ft_pct_home' por 0
game_filtrada['ft_pct_home'] = game_filtrada['ft_pct_home'].fillna(0)

game_filtrada
```

```python
game_seasontype=pd.unique(game_filtrada["plus_minus_away"])
game_seasontype
```

```python
game_seasontype=pd.unique(game_filtrada["season_type"])
game_seasontype
```

```python
#selecciono y filtro solo los partidos de temporada regular y playoffs
game_filtrada = game_filtrada[~game_filtrada['season_type'].isin(['All-Star', 'All Star', 'Pre Season'])]

game_filtrada
```
