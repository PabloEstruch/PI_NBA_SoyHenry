```python
data = pd.read_csv("common_player_info.csv", sep=',')

print(data.columns)
```

```python
columnas_a_eliminar = ['display_first_last', 'display_last_comma_first', 'display_fi_last', 'player_slug', 'last_affiliation', 'team_id', 'team_code', 'dleague_flag', 'nba_flag', 'games_played_flag', 'greatest_75_flag', 'games_played_current_season_flag']
data_cleaned = data.drop(columns=columnas_a_eliminar)

#playercode no creo que sirva, pero lo dejo por si acaso

#Para ver todas las columnas
pd.set_option('display.max_columns', None)
data_cleaned
```

```python
data_cleaned.to_csv("common_player_info_clean.csv", index=False)
```
