```python
# Importo librerias

import pandas as pd
```

```python
data = pd.read_csv("draft_combine_stats.csv", sep=',')

print(data.columns)
```

```python
data
```

```python
columnas_a_eliminar = ['player_name', 'height_wo_shoes', 'height_w_shoes', 'height_w_shoes_ft_in', 'wingspan', 'standing_reach', 'spot_fifteen_corner_left', 'spot_fifteen_break_left', 'spot_fifteen_top_key', 'spot_fifteen_break_right', 'spot_fifteen_corner_right', 'spot_college_corner_left', 'spot_college_top_key', 'spot_college_break_right', 'spot_college_corner_right', 'spot_nba_corner_left', 'spot_nba_break_left', 'spot_nba_top_key', 'spot_nba_break_right', 'spot_nba_corner_right', 'off_drib_fifteen_break_left', 'off_drib_fifteen_top_key', 'off_drib_fifteen_break_right', 'off_drib_college_break_left', 'on_move_college', 'on_move_fifteen', 'off_drib_college_break_right', 'off_drib_college_top_key', 'spot_college_break_left']
data_cleaned = data.drop(columns=columnas_a_eliminar)



#Para ver todas las columnas
pd.set_option('display.max_columns', None)
data_cleaned
```

```python
# Esto lo use para verificar todas las columnas antes de eliminarlas


# Contar los valores nulos en una columna específica
nulos_columna = data_cleaned['bench_press'].isnull().sum()

# Mostrar el número de valores nulos
print(nulos_columna)
```

```python
# Promediar las dos columnas, ignorando los valores nulos de la segunda columna
data_cleaned['lane_agility_time_complete'] = data_cleaned[['lane_agility_time', 'modified_lane_agility_time']].mean(axis=1)

pd.set_option('display.max_columns', None)
data_cleaned
```

```python
data_cleaned = data_cleaned.drop(columns=['lane_agility_time', 'modified_lane_agility_time'])
data_cleaned
```

```python
data_filtered = data_cleaned.query('2001 <= season <= 2023')
data_filtered
```

```python
# Esto lo use para verificar todas las columnas antes de eliminarlas


# Contar los valores nulos en una columna específica
nulos_columna = data_filtered['position'].isnull().sum()

# Mostrar el número de valores nulos
print(nulos_columna)
```

```python
data_filtered.to_csv("draft_combine_stats_clean.csv", index=False)
```
