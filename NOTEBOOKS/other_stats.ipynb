Revisar la tabla de game_filtrada para filtrar la informacion respecto a los juegos de analisis

```python
game_ids_filtrados = game['game_id'].unique().tolist()
game_ids_filtrados
```

```python
len(game_ids_filtrados)
```

Voy a filtrar el df con los game_id filtrados por año

```python
df_filtrado = df[df['game_id'].isin(game_ids_filtrados)]
df_filtrado
```

```python
duplicados_filtrados = df_filtrado[df_filtrado.duplicated(keep=False)]
print(duplicados_filtrados)
```

```python
# Convertir las columnas a cadenas de texto
df_filtrado['team_city_home'] = df_filtrado['team_city_home'].astype(str)
df_filtrado['team_city_away'] = df_filtrado['team_city_away'].astype(str)
contiene_all_stars_home = df_filtrado['team_city_home'].str.contains('All Stars', case=False, na=False).any()
contiene_all_stars_away = df_filtrado['team_city_away'].str.contains('All Stars', case=False, na=False).any()
if contiene_all_stars_home or contiene_all_stars_away:
    print("La frase 'All Stars' está presente en al menos una fila de las columnas 'team_city_home' o 'team_city_away'.")
else:
    print("La frase 'All Stars' no está presente en ninguna fila de las columnas 'team_city_home' o 'team_city_away'.")
```

```python
filas_con_all_stars_home = df_filtrado[df_filtrado['team_city_home'].str.contains('All Stars', case=False, na=False)]
filas_con_all_stars_away = df_filtrado[df_filtrado['team_city_away'].str.contains('All Stars', case=False, na=False)]
```

Analisis de los puntos y eficiencia ofensiva

```python
df_filtrado.groupby("team_abbreviation_home")[["pts_paint_home", "pts_2nd_chance_home", "pts_fb_home", "pts_off_to_home"]].mean().sort_values("pts_paint_home", ascending=False)
```

```python
df_filtrado.groupby("team_abbreviation_away")[["pts_paint_away", "pts_2nd_chance_away", "pts_fb_away", "pts_off_to_away"]].mean().sort_values("pts_paint_away", ascending=False)
```

Analisis de rebotes y perdidas del balon

```python
df_filtrado.groupby("team_abbreviation_away")[["team_rebounds_away", "total_turnovers_away"]].mean().sort_values("team_rebounds_away", ascending=False)
```
