```python
# Importo librerias

import pandas as pd
```

```python
df_draft_history = pd.read_csv(r"../Data/draft_history.csv", sep=',')

print(df_draft_history.columns)
```

```python
#Filtrar por fecha
df_draft_history = df_draft_history[(df_draft_history["season"] >= 2001) & (df_draft_history["season"] <= 2023)]
```

```python
df_draft_history.drop(columns=[ 'player_name', 'season', 'round_number', 'round_pick',
       'overall_pick', 'draft_type', 'team_city', 'team_name',
       'team_abbreviation', 'organization', 'organization_type',
       'player_profile_flag'], inplace=True)
```
