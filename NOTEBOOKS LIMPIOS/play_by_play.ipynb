```python
import numpy as np
import pandas as pd
```

```python
#lectura de los datos
df_play_by = pd.read_csv(r'../Data/play_by_play.csv')
```

```python
#lectura de los datos de game
df_game = pd.read_csv(r"../Data/game_copy.csv", usecols=["game_id"])
```

```python
df_play_by.columns
```

```python
#Eliminar columnas
columnas_a_eliminar = ['eventnum', 'eventmsgactiontype', 'period',
       'wctimestring', 'pctimestring', 'homedescription', 'neutraldescription',
       'visitordescription', 'scoremargin', 'person1type',
       'player1_name', 'player1_team_id', 'player1_team_city',
       'player1_team_nickname', 'player1_team_abbreviation', 'person2type',
       'player2_name', 'player2_team_id', 'player2_team_city',
       'player2_team_nickname', 'player2_team_abbreviation', 'person3type',
       'player3_id', 'player3_name', 'player3_team_id', 'player3_team_city',
       'player3_team_nickname', 'player3_team_abbreviation',
       'video_available_flag']
# Eliminar las columnas
df_play_by = df_play_by.drop(columns=columnas_a_eliminar)
```

```python
#Lista de valores unicos de df_game
df_game_ids_filtrados = df_game['game_id'].unique().tolist()
```

```python
#Filtrar play_by_play por game_id
df_play_by = df_play_by[df_play_by['game_id'].isin(df_game_ids_filtrados)]
```

```python
#Filtrar play_by_play por eventmsgtype = 1,2,3,4,5,6
eventos_analisis= [1, 2, 3, 4, 5, 6]
df_play_by = df_play_by[df_play_by['eventmsgtype'].isin(eventos_analisis)]
```

```python
#Generar Booleanos en la columna score para contabilizar tiros libres convertidos y fallados
df_play_by['score'] = np.where(df_play_by['score'].isna(), 0, 1)
```

```python
#Filtrar la columna player1_id por valores menores a 2.000.000
df_play_by =df_play_by[df_play_by['player1_id'] < 2000000]
```
